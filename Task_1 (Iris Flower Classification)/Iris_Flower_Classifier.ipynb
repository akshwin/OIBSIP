```python
df = df.drop('Id',axis=1)
```

```python
x = df.drop('Species',axis = 1)
```

```python
y = df['Species']
```

```python
y.head()
```

**DIVIDING INTO TRAINING AND TESTING DATA**

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 1)
```

**BUILDING THE MODEL**

```python
from sklearn import svm 

model = svm.SVC(C=3)
```

```python
model.fit(x_train,y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
y_predict
```

```python
model.score(x_test,y_test)
```

```python
y_grid = (np.column_stack([y_test,y_predict]))
```

```python
y_grid
```

```python
import string 
lab = list(['Iris-setosa','Iris-virginica','Iris-versicolor'])
plab = ["Predicted " + s for s in lab]
```

```python
lab
```

```python
from sklearn import metrics

cm = metrics.confusion_matrix(y_test,y_predict)

df_cm = pd.DataFrame(cm , index = [ i for i in lab],
                     columns = [ i for i in plab])
```

```python
plt.figure(figsize = (20,13))
sns.heatmap(df_cm , annot = True , fmt='g',cmap='PiYG')
plt.show()
```
